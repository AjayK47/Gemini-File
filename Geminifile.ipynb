### Import Libraries Required

```python
from llama_index import SimpleDirectoryReader, VectorStoreIndex
from llama_index.llms import Gemini
from IPython.display import Markdown, display
from llama_index import ServiceContext
from llama_index import StorageContext, load_index_from_storage
import os
```

### Make a Folder named 'data' and upload all the necessary pdf files you want to excute in that folder

```python
documents = SimpleDirectoryReader("data").load_data()
```

```python
# intailise the model
llm = Gemini()
```

There are Different types of Embeddding models, for Llama index you will be default given Open AI embedding for that you need you open AI key and it is not open source , so it is preffered to use open source models , you can find them from Hugging face community.
Here what i am using is UAE-Large-V1 model at the time of making this project this model is currently top of hugging face leaderboard for text embedding models, you can find the leaderboard here https://huggingface.co/spaces/mteb/leaderboard and select the best possible model for you and just change the name of model in below code.

### Load your HuggingFace Embeddig model

```python
from llama_index.embeddings import HuggingFaceEmbedding
embed_model_uae = HuggingFaceEmbedding(model_name="WhereIsAI/UAE-Large-V1")
```

```python
# Configure Service Context
service_context = ServiceContext.from_defaults(llm=llm, chunk_size=800, chunk_overlap=20,embed_model=embed_model_uae)
```

#### Indexing your data

```python
# This step may take some-time depending of size of data and system you are using

index = VectorStoreIndex.from_documents(documents,service_context=service_context,show_progress=True)
```

```python
# Persisting data to temporary store index and not make one for every query
index.storage_context.persist()
```

```python
# Queries your Data from made up index database of your document
query_engine = index.as_query_engine()

# Perform a query on the indexed documents
response = query_engine.query("what are recent developments in india")
```

```python
# Display the query response using Markdown with bold formatting

display(Markdown(f"<b>{response}</b>"))
```
